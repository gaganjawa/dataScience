```python
import numpy as np
```

```python
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
sns.countplot(x = 'sex', data=df)
```

```python
sns.countplot(x  = 'smoker', data=df)
```

```python
sns.countplot(x='smoker', data=df, hue='time')
```

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

```python
sns.barplot(x='sex', y='total_bill', data=df, hue='smoker')
```

```python
sns.boxplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data=df, hue='sex')
```

```python
sns.boxenplot(x='day', y='total_bill', data=df)
```

```python
sns.violinplot(x='day', y='total_bill', data=df)
```

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=df)
```

```python
sns.swarmplot(x='day', y='total_bill', data=df, hue='sex')
```
