```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
a = np.arange(1,11)
```

```python
a
```

```python
b = a**2
```

```python
plt.plot(a,b)
```

```python
plt.plot(a,b,'x')
plt.xlabel('Numbers')
plt.ylabel('Squares')
```

```python
plt.plot(a,b,'rx-')
```

```python
plt.subplot(1,2,1)
plt.plot(a,b,'rx-')
plt.subplot(1,2,2)
plt.plot(b,a)
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes.set_title('A1')
axes.set_xlabel('numbers')
axes.set_ylabel('square')

axes2.set_title('A2')
axes2.set_xlabel('Square')
axes2.set_ylabel('Numbers')

axes.plot(a,b)
axes2.plot(b,a)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,0.4,1])
axes2 = fig.add_axes([0.5,0,0.4,1])

axes.plot(a,b)
axes2.plot(b,a)
```

```python
fig.savefig("mypicture.png")
```
